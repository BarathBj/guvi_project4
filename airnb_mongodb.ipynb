```python
#Connections
import pymongo
import pandas as pd
pd.set_option("display.max_columns",None)
import warnings
warnings.filterwarnings("ignore")
```

```python
#MongoDB
client = pymongo.MongoClient("mongodb+srv://Barath:12345@cluster0.xbpqods.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0")
db = client["sample_airbnb"]
collection = db["listingsAndReviews"]
```

```python
document = collection.find({}, {'_id': 0})
document[0]
```

```python
cols= { 'name':[],'property_type':[], 'price':[], 'security_deposit':[], 'cleaning_fee':[], 'extra_people_fee':[], 'guests_included':[],
     'minimum_nights':[], 'cancellation_policy':[], 'accommodates':[], 'bedrooms':[], 'beds':[], 'bed_type':[], 'bathrooms':[],
     'country':[], 'number_of_reviews':[], 'review_scores_rating':[] }

for i in collection.find():
  name = i.get('name')
  property_type = i.get('property_type')
  price = i.get('price')

  security_deposit = i.get('security_deposit')
  cleaning_fee = i.get('cleaning_fee')
  extra_people_fee = i.get('extra_people')
  guests_included = i.get('guests_included')

  minimum_nights = i.get('minimum_nights')
  cancellation_policy = i.get('cancellation_policy')
  accommodates = i.get('accommodates')
  bedrooms = i.get('bedrooms')
  beds = i.get('beds')

  bed_type = i.get('bed_type')
  bathrooms = i.get('bathrooms')

  country = i['address'].get('country')
  number_of_reviews = i.get('number_of_reviews')
  review_scores_rating = i['review_scores'].get('review_scores_rating')


  cols['name'].append(name)
  cols['property_type'].append(property_type)
  cols['price'].append(price)

  cols['security_deposit'].append(security_deposit)
  cols['cleaning_fee'].append(cleaning_fee)
  cols['extra_people_fee'].append(extra_people_fee)
  cols['guests_included'].append(guests_included)

  cols['minimum_nights'].append(minimum_nights)
  cols['cancellation_policy'].append(cancellation_policy)
  cols['accommodates'].append(accommodates)
  cols['bedrooms'].append(bedrooms)
  cols['beds'].append(beds)

  cols['bed_type'].append(bed_type)
  cols['bathrooms'].append(bathrooms)

  cols['country'].append(country)
  cols['number_of_reviews'].append(number_of_reviews)
  cols['review_scores_rating'].append(review_scores_rating)


df = pd.DataFrame(cols)
```
